```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image  # Import the Image module from the PIL (Pillow) library
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # Import ImageDataGenerator
import os  # Import the os module

# Load a single image
cat_img = Image.open(r"D:/Data Augmantation/augmented_images/chair.jpg")

# Convert the image to a NumPy array
cat_img = np.array(cat_img)
print('Shape of Image before editing:', cat_img.shape)  # Check the original image shape

# We want to add a dimension for batch
cat_img = cat_img[np.newaxis, ...]
print('Shape of Image after editing:', cat_img.shape)

# Initialize the ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Folder path to save augmented images
output_folder = "D:/Data Augmantation/good chair/new2"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Generate augmented images and save them to the output folder
i = 0
for batch in datagen.flow(cat_img, batch_size=1):
    augmented_image = batch[0] / 255.0  # Normalize the pixel values to [0.0, 1.0]
    
    # Save the augmented image to the output folder
    image_filename = os.path.join(output_folder, f"augmented_chair_{i}.jpg")
    plt.imsave(image_filename, augmented_image)
    
    plt.figure(i)
    plt.imshow(augmented_image)
    i += 1
    if i % 4 == 0:
        break  # Stop after displaying and saving 4 augmented images

plt.show()
```
